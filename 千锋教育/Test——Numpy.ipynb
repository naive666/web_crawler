```python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
x=np.array([[0,0.5,1],[0,0.5,1]])
y=np.array([[0,0,0],[1,1,1]])
print('the dimension of x is {}, the shape is {}'.format(x.ndim,x.shape))
print('the dimension of y is {}, the shape is {}'.format(y.ndim,y.shape))

plt.plot(x,y,'.')
contor=np.sqrt(x**2+y**2)
plt.plot(contor)
plt.grid(True)
plt.show()
```

```python
plt.imshow(contor)
plt.colorbar()
plt.show()
```

```python
##若要表示很多个点，则可以用meshgrid()
##现在想要生成一个20行，30列的网络点
x=np.arange(10)
y=np.linspace(0,60,15)

r,c=np.meshgrid(x,y)
plt.plot(r,c,'.')
plt.grid(True)
plt.show()
```

```python
con=np.sqrt(r**2+c**2)
plt.plot(con)
plt.grid(True)
plt.show()
```

```python
plt.imshow(con)
plt.colorbar()
plt.show()
```
