```python
import urllib.request
import urllib.parse
import os
from bs4 import BeautifulSoup
import time
```

```python
class qiushi(object):
    url='https://www.qiushibaike.com/text/page/'
    
    def __init__(self,start_page,end_page):
        self.start_page=start_page
        self.end_page=end_page
        self.item0=[]
        self.item1=[]
        self.item2=[]
        
        
        
    def run(self):
        for page in range(self.start_page,self.end_page+1):
            request=self.handle_request(page)
            content=urllib.request.urlopen(request).read().decode()
            self.parse_content(content)
            time.sleep(1)
            filename='qiushi'+str(page)+'.txt'
            #with open(filename,'w') as fp:
            print(self.item0)
            print(self.item1)
            print(self.item2)
    
    def parse_content(self,content):
        soup=BeautifulSoup(content,'lxml')
        
        long_article=soup.find_all('div',class_='article block untagged mb15 typs_long')
        for my_long in long_article:
            text_long=my_long.find('div',class_='content').text
            haoxiao0_long=my_long.find_all('i',class_='number')[0].text
            comment_long=my_long.find_all('i',class_='number')[1].text
            main_com_long=my_long.find('div',class_='main-text').text
            item_long={
                '内容':text_long,
                '好笑':haoxiao0_long,
                '评论':comment_long,
                '评论内容':main_com_long
            }
            self.item0.append(item_long)
        
        hot_article=soup.find_all('div',class_='article block untagged mb15 typs_hot')
        for my_hot in hot_article:
            text_hot=my_hot.find('div',class_='content').text
            haoxiao0_hot=my_hot.find_all('i',class_='number')[0].text
            comment_hot=my_hot.find_all('i',class_='number')[1].text
            main_com_hot=my_hot.find('div',class_='main-text').text
            item_hot={
                '内容':text_hot,
                '好笑':haoxiao0_hot,
                '评论':comment_hot,
                '评论内容':main_com_hot
            }
            self.item1.append(item_hot)
        
        old_article=soup.find_all('div',class_='article block untagged mb15 typs_old')
        for my_old in old_article:
            text_old=my_old.find('div',class_='content').text
            haoxiao0_old=my_old.find_all('i',class_='number')[0].text
            comment_old=my_old.find_all('i',class_='number')[1].text
            main_com_old=my_old.find('div',class_='main-text').text
            item_old={
                '内容':text_old,
                '好笑':haoxiao0_old,
                '评论':comment_old,
                '评论内容':main_com_old
            }
            self.item2.append(item_old)
    
    def handle_request(self,page):
        url=self.url+str(page)+'/'
        headers={
            'Connection': 'keep-alive',
            'Cache-Control': 'max-age=0',
            'Upgrade-Insecure-Requests': '1',
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/63.0.3239.132 Safari/537.36',
            'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8',
            'Accept-Language': 'zh-CN,zh;q=0.9',
            'Cookie': '_xsrf=2|4c92672c|e0537f7e88e3a33aee81798ac90b6544|1551620868; Hm_lvt_2670efbdd59c7e3ed3749b458cafaa37=1551364903,1551365517,1551366428,1551620870; _ga=GA1.2.84505463.1551620872; _gid=GA1.2.1227993367.1551620872; __cur_art_index=9000; Hm_lpvt_2670efbdd59c7e3ed3749b458cafaa37=1551624578',
            'If-None-Match': '175627d3a25d8af864cf34e171df31e0ffe1e3c4'
}
        
        request=urllib.request.Request(url=url,headers=headers)
        return request


def main():
    start_page=int(input('Please enter the start page:'))
    end_page=int(input('Please enter the end page:'))
    
    spider=qiushi(start_page,end_page)
    spider.run()

    
    
if __name__=='__main__':
    main()
```
